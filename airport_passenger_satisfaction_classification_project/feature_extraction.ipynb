```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
```

```python
data_train=pd.read_csv("data/train.csv")
data_test=pd.read_csv("data/test.csv")
```

```python
data_train.head()
```

```python
data_train["is_train"]=1
data_test["is_train"]=0
df=pd.concat([data_train,data_test],ignore_index=True)
```

```python
df.head()
```

```python
df.drop(columns={"Unnamed: 0","id"},axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean())
df.head()
```

```python
categorical_data=df.nunique() ==2
col=df.columns[categorical_data]
col
```

```python
label_encoder=LabelEncoder()
for i in range(4):
    df[col[i]]=label_encoder.fit_transform(df[col[i]])
```

```python
df=pd.get_dummies(df,columns=["Class"])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.nunique()
```

```python
scaler_list=["Age","Flight Distance","Departure Delay in Minutes","Arrival Delay in Minutes"]
```

```python
scaler=MinMaxScaler()
df[scaler_list]=scaler.fit_transform(df[scaler_list])
```
